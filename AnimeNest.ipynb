```python
# List of columns to drop
columns_to_drop = ['English name', 'Other name', 'Aired','Duration', 
                   'Rating', 'Rank','Licensors', 'Popularity', 'Favorites', 'Scored By', 
                   'Members', 'Image URL']

# Drop columns
anime.drop(columns=columns_to_drop, inplace=True)

# Display modified dataframe
print(anime.head())
```

```python
# Convert genre column to lowercase and remove leading/trailing whitespaces
anime['Genres'] = anime['Genres'].str.lower().str.strip()

# Reset index after dropping rows
anime = anime.reset_index(drop=True)
```

```python
anime['tags'] = anime['Genres'] + anime['Synopsis'] + anime['Producers'] + anime['Studios'] + anime['Source']
```

```python
df=anime[['anime_id','Name','Synopsis','tags']]
```

```python
# Convert synopsis to lowercase and remove punctuation
df['Synopsis'] = df['Synopsis'].str.lower()
df['Synopsis'] = df['Synopsis'].str.replace('[^\w\s]', '')

# Handle missing values if any
df['Synopsis'].fillna('', inplace=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english')

# Fit and transform the data
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Synopsis'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def get_recommendations(title, cosine_sim, indices, df, top_n=10):
    # Get index of anime title
    idx = indices[title]
    
    # Get pairwise similarity scores with all other anime
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort anime based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get top N most similar anime (excluding itself)
    top_similar_anime = sim_scores[1:top_n+1]
    
    # Get indices of recommended anime
    recommended_indices = [anime[0] for anime in top_similar_anime]
    
    # Return titles of recommended anime
    return df['Name'].iloc[recommended_indices]
```

```python
# Create a mapping of anime titles to their indices in the dataset
indices = pd.Series(df.index, index=df['Name']).drop_duplicates()
```

```python
def recommend_anime():
    # Input anime title from user
    title = input("Enter the name of an anime: ").strip()
    
    # Check if the anime title exists in the dataset
    if title in indices:
        recommendations = get_recommendations(title, cosine_sim, indices, df)
        print(f"Recommendations for '{title}':")
        for idx, recommendation in enumerate(recommendations, start=1):
            print(f"{idx}. {recommendation}")
    else:
        print(f"Anime '{title}' not found in the dataset. Please try again.")
```

```python
recommend_anime()
```
